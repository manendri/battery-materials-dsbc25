```python
import pandas as pd
```

#### load the training dataset
This dataframe contains a row for each of the 8648 MOFs in the training set and a subset of columns from the original qmof.csv dataset.

```python
df_train = pd.read_csv('../../data/datasets_main/train_hse06.csv', sep=',')
df_train
```

#### load one of the fingerprints datasets
This dataframe contains one of the extra sets of features (here 'stoich45'), with a row for every MOF in the entire qmof database. Note: we should _not_ use all of these rows for feature selection because:
1) they include MOFs which don't have HSE06 bandgap values.
2) they include MOFs in the test set.

```python
df_stoich45 = pd.read_csv('../../data/QMOF/qmof_database/qmof_database/relaxed_structures_fingerprints/stoich45_fingerprints.csv', sep=',')
df_stoich45
```

#### merge the two dataframes above
Merging puts the columns / potential features from both dataframes into one dataframe, on which we can do EDA and feature selection.

df_train lists the QMOF id in the 'qmof_id' column, but df_stoich45 lists it under 'MOF'. By using the `left_on = 'MOF'` and `right_on = 'qmof_id'` options in `pd.merge` below, the rows in df_train and df_stoich45 are matched up by QMOF id. By using `how = 'right'`, only those 8648 MOFs from df_train appear in the resulting dataframe.

```python
# Merge the DataFrames on the matching columns
df_train_stoich45 = pd.merge(df_stoich45, df_train, left_on='MOF', right_on='qmof_id', how='right')
df_train_stoich45
```

Note: the new dataframe includes both a 'MOF' column and a 'qmof_id' column, but they are redundant.

```python
df_train_stoich45[['MOF','qmof_id']]
```

#### adding one more set of features to the dataframe

```python
#import another 'fingerprint' dataset
df_stoich120 = pd.read_csv('../../data/QMOF/qmof_database/qmof_database/relaxed_structures_fingerprints/stoich120_fingerprints.csv', sep=',')
df_stoich120
```

```python
# merge the stoich120 dataframe into df_train_stoich45
df_train_stoich45.merge(df_stoich120)
```

Since both df_train_stoich45 and df_stoich120 list the QMOF id under 'MOF' there was no need to include the additional arguments in merge() here.
