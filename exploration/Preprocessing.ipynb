## Handling Nan values

Suggested approach for handling Nan values:

1. Drop columns with most nan values 
2. Drop rows with most nan values
3. Fill remaining missing values using imputation

Lets check which columns do we have in the dataset

```python
print('Total no. of columns:', len(df.columns))
df.columns
```

#### It would make sense to remove some redundant columns at this stage. 

Which columns are most important here?


```python
columns_to_remove = [
    'name', 'info.formula', 'info.formula_reduced',
       'info.mofid.mofid', 'info.mofid.mofkey', 'info.mofid.smiles_nodes',
       'info.mofid.smiles_linkers', 'info.mofid.smiles', 'inputs.pbe.theory',
       'inputs.pbe.pseudopotentials', 'inputs.pbe.encut', 'inputs.pbe.kpoints',
       'inputs.pbe.gamma', 'inputs.hse06.theory',
       'inputs.hse06.pseudopotentials', 'inputs.hse06.encut', 'inputs.hse06.kpoints',
       'inputs.hse06.gamma', 'inputs.hle17.theory',
       'inputs.hle17.pseudopotentials', 'inputs.hle17.encut', 'inputs.hle17.kpoints',
       'inputs.hle17.gamma', 'inputs.hse06_10hf.theory',
       'inputs.hse06_10hf.pseudopotentials', 'inputs.hse06_10hf.encut', 'inputs.hse06_10hf.kpoints',
       'inputs.hse06_10hf.gamma'
] # Please update based on the columns you want to remove

df.drop(columns_to_remove, axis=1, inplace=True)
print(len(df.columns), 'columns remaining')
```

Understand the distribution of nan values

```python
# How are the nan values distributed in the dataset?
missing_summary = df.isna().sum()
print(missing_summary[missing_summary > 0])  # Show columns with NaN values
```

```python
# Drop Columns with Too Many NaNs (e.g., more than 50%) - may need to modify NaN treatment (i. e. keep pbe and hse06 columns only)

threshold = 0.5 * len(df)
df_cleaned_cols = df.dropna(thresh=threshold, axis=1)
print('Total no. of columns after dropping columns with too many NaNs:', len(df_cleaned_cols.columns), 'out of', len(df.columns))
df_cleaned_cols
```

```python
# Drop rows/mofs with too many NaNs (e.g., more than 5)

df_cleaned_rows = df_cleaned_cols.dropna(thresh=50, axis=0)  # Keep rows with at least 50 non-null values (Rudimentary, more fine tuning needed)
print('Total no. of rows after dropping rows with too many NaNs:', len(df_cleaned_rows), 'out of', len(df))
df_cleaned_rows
```
