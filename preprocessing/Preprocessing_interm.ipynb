### Definition of column names (obtained from QMOF'22 descriptions):
For a given entry in `qmof`, the following tabulated properties can be accessed:

`qmof_id`: - A unique identifier for the material (str).

`name`: - The internally used filename (str).

`info`: - `formula`: The chemical formula per unit cell (str). - `formula_reduced`: The reduced (i.e. empirical) formula (str). - `mofid`: The MOFid information. - `mofid`: The MOFid (str). - `mofkey`: The MOFkey (str).. - `smiles_nodes`: SMILES representation of the nodes (list of str). - `smiles_linkers`: SMILES representation of the linkers (list of str). - `smiles`: Concatenated SMILES of the nodes and linkers (str). - `topology`: The topology code (str). - `natoms`: The number of atoms per unit cell (int). - `pld`: The pore-limiting diameter computed using Zeo++, in Å (float). - `lcd`: The largest cavity diameter computed using Zeo++, in Å (float). - `density`: The mass-density, in g/cm^3 (float). - `symmetry`: The symmetry information. - `spacegroup_number`: The spacegroup number. - `spacegroup_crystal`: The crystal system. - `pointgroup`: The point group. - `synthesized`: Whether the material is from an experimental MOF database (bool). - `source`: The source of the initial crystal structure (str). - `doi`: The URL pointing to the source of the crystal structure (str).

`inputs`:
`pbe`: Calculations carried out on PBE-D3(BJ) optimized geometries with properties presented at the PBE-D3(BJ) level of theory. - `theory`: Level of theory (str). - `pseudopotentials`: VASP 5.4 pseudopotentials used (list of str) - `encut`: The VASP ENCUT parameter (int). - `kpoints`: The k-points used in the calculation (list of ints). - `gamma`: If a gamma-centered k-point grid is specified in the KPOINTS file (bool). - `spin`: If the calculation is spin-polarized (bool).
`hle17`: Calculations carried out on PBE-D3(BJ) optimized geometries with properties presented at the HLE17 level of theory. Only present for some materials. - Same as `pbe`.
`hse06_10hf`: Calculations carried out on PBE-D3(BJ) optimized geometries with properties presented at the HSE06\* level of theory. Only present for some materials. - Same as `pbe`.
`hse06`: Calculations carried out on PBE-D3(BJ) optimized geometries with properties presented at the HSE06 level of theory. Only present for some materials. - Same as `pbe`.

`outputs`:
`pbe`: Calculations carried out on PBE-D3(BJ) optimized geometries with properties presented at the PBE-D3(BJ) level of theory. - `energy_total`: The DFT-computed energy in eV (float). - `energy_vdw`: Only the D3(BJ) contribution to the energy in eV (float). - `energy_elec`: Only the electronic contribution to the energy in eV (float). - `net_magmom`: Absolute value of the overall magnetic moment (float). - `bandgap`: The band gap in eV computed from the Kohn-Sham eigenvalues (float). - `cbm`: The conduction band minimum in eV computed from the Kohn-Sham eigenvalues (float). - `vbm`: The valence band maximum in eV computed from the Kohn-Sham eigenvalues (float). - `directgap`: Whether the band gap is direct, as opposed to indirect (bool). - `bandgap_spins`: The band gap for the up- and down-spin channels (list of floats). - `cbm_spins`: The CBM for the up- and down-spin channels (list of floats). - `vbm_spins`: The VBM for the up- and down-spin channels (list of floats). - `directgap_spins`. Whether the band gap in the up- and down-spin channels is direct, as opposed to indirect (list of bools).
`hle17`: Calculations carried out on PBE-D3(BJ) optimized geometries with properties presented at the HLE17 level of theory. Only present for some materials. - Same as `pbe`.
`hse06_10hf`: Calculations carried out on PBE-D3(BJ) optimized geometries with properties presented at the HSE06\* level of theory. Only present for some materials. - Same as `pbe`.
`hse06`: Calculations carried out on PBE-D3(BJ) optimized geometries with properties presented at the HSE06 level of theory. Only present for some materials. - Same as `pbe`.

#### Pattern of NaNs
Are the missing entries structured in some way? How many different types of 'profiles' of missing entries across rows appear in the dataset? And how many rows have each profile type?

From a quick visual inspection of the data: if a row has missing entries, they seem to be missing in 'chunks', in which entries all appear or are all missing. It turns out (as is shown below) that the chunks correspond the different DFT calculations, which is made clear by looking at the column names. We start by spliting the columns into two categories (info and DFT) and extracting the prefix of each column:

```python
# separate 'info' columns from 'DFT input/output' columns
info_column_names = [cn for cn in df.columns if cn == 'qmof_id' or cn == 'name' or cn.startswith('info')]
dft_column_names = [cn for cn in df.columns if cn.startswith('inputs') or cn.startswith('outputs')]

# Do info_column_names and dft_column_names cover all columns?
set(info_column_names + dft_column_names) == set(df.columns)
```

```python
# define a function to get the prefix of column name, i.e. part of the column name before the last '.' or the whole column name if it contains no '.'
def get_column_prefix(cn):
    if '.' in cn:
        return '.'.join(cn.split('.')[:-1])
    else:
        return cn

# list of unique DFT column  prefixes
dft_column_prefixes = list(dict.fromkeys([get_column_prefix(cn) for cn in dft_column_names]))
dft_column_prefixes
```

The image plot below represents all 'profiles' of the rows of the original data frame, restricted to the DFT (i.e., 'inputs.\*' and 'outputs,\*') columns. Each row in the image shows the number of data rows with that profile and the profile itself. The color of each entry in the profile is either red (if those data rows have that entry missing) or colored according to the column prefix (if the entry is present).

Importantly, this image shows that the dataset is not randomly / spordaically missing entries. Rather, for some MOFs, only a subset of the DFT calculations were completed. (In fact there are 2^3 = 8 different profiles, which is the number of subsets of a set of 3 members - the 'pbe' cacluation was always done, but the other 3 DFT calculation may or may not have been done, and all combinations appear in the dataset.) 


```python
# a data frame with the same shape as the original, restricted to the DFT columns, but with entries replaced with an index: 0 if the entry was NaN or a positive integer representing the column prefix if not
notna_column_prefix_id_df = pd.DataFrame(
   np.array([dft_column_prefixes.index(get_column_prefix(cn)) + 1 for cn in dft_column_names]).reshape(1,len(dft_column_names)).repeat(df.shape[0],axis=0),
    columns = dft_column_names
) * df[dft_column_names].notna()

# tally of idential rows in the above data frame (note: the tally is in the 'size' column of this data frame)
tally_inout_df = notna_column_prefix_id_df.groupby(dft_column_names,as_index=False).size().sort_values(by='size',ascending=False)

# plot the tally data frame as a image
plt.figure(figsize=(10,5))

plt.imshow(np.array(tally_inout_df)[:,:-1], cmap = 'Set1')
plt.xlim(left=-25)
for i, tally in enumerate(tally_inout_df['size']):
    plt.text(-23,i,'number of rows: ' + str(tally), verticalalignment = 'center')

ax = plt.gca()
ax.set_aspect(5)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xticks(np.arange(-.5,len(dft_column_names)+.5, 1), minor=True)
ax.set_yticks(np.arange(.5,7.5, 1), minor=True)
ax.tick_params(axis='both',which='minor',length=0)
plt.grid(visible=True, color='black', linewidth = .4, axis='both', which = 'minor')

cbar = plt.colorbar()
cbar.ax.set_yticks(np.arange(0, len(dft_column_prefixes)+1, 1))
cbar.ax.set_yticklabels(['missing entry'] + ['column = '+prefix+'.*' for prefix in dft_column_prefixes])

plt.show
```

# Let's get ready to output processed version/s of the dataset

We would consider two ways in processing our dataset:

1. Separate into four subsets for each level of theory (PBE, HLE17, HSE06, HSE06*), as needed for training individual models
2. Consider each MOF at each level of theory as an individual data point and generate a reformatted dataset as needed for training multi-fidelity models

#### It would make sense to remove some redundant columns at this stage. 

Which columns are most important here?


Any structural information (info.) containing SMILES code retained. All columns related to DFT simulation details (input.) dropped.

```python
# Define a function that drops the DFT level of theory part of the inputs/outputs column names, leaving other (info) column names untouched
# except for ones with the strict format "outputs.*.bandgap" and "inputs.theory" in the column name
def drop_level_from_column_name(col):
    if col.startswith('outputs') and col.endswith('.bandgap'):
        return col
    elif 'inputs.theory' in col:
        return col
    elif col.startswith('inputs') or col.startswith('outputs'):
        return '.'.join(col.split('.')[::2])
    else:
        return col
    
# Apply the drop_level_from_column_name function to rename the columns
df.columns = [drop_level_from_column_name(col) for col in df.columns]

# Keep only 'inputs.theory', columns of the strict format 'outputs.*.bandgap', 'info.*', and 'name' columns
columns_to_keep = ['inputs.theory'] + [col for col in df.columns if col.startswith('outputs') and col.endswith('.bandgap')] + [col for col in df.columns if col.startswith('info')] + [col for col in df.columns if 'name' in col] + [col for col in df.columns if 'qmof_id' in col]
df = df[columns_to_keep]

# Remove irrelevant columns based on domain knowledge
df.drop([
    'name', 'info.formula', 'info.formula_reduced', 'info.mofid.mofid', 'info.mofid.mofkey', 'info.mofid.topology', 'info.symmetry.spacegroup',
    'info.symmetry.spacegroup_number', 'info.symmetry.spacegroup_crystal', 'info.symmetry.pointgroup', 'info.synthesized', 'info.source', 'info.doi'
], axis=1, inplace=True)


# Modify the values in 'inputs.theory' columns using the below notation:
# PBE-D3BJ -> PBE
# HLE17//PBE-D3BJ -> HLE17
# HSE06-D3BJ-HF10//PBE-D3BJ -> HSE06*
# HSE06-D3BJ//PBE-D3BJ -> HSE06
# This is done to make the values more readable and match the QMOF'22 paper
df.replace(to_replace={'PBE-D3BJ': 'PBE', 'HLE17//PBE-D3BJ': 'HLE17', 'HSE06-D3BJ-HF10//PBE-D3BJ': 'HSE06*', 'HSE06-D3BJ//PBE-D3BJ': 'HSE06'}, inplace=True)

# Display the modified dataframe
df
```

**Method 1.** Separate into four subsets for each level of theory (PBE, HLE17, HSE06, HSE06*), as needed for training individual models

some name formatting for 'inputs.theory' type columns:

```python
# Rename duplicate 'inputs.theory' columns to 'inputs.theory.0', 'inputs.theory.1', etc.
from collections import Counter

def rename_duplicates(cols):
    counter = Counter()
    new_cols = []
    for col in cols:
        cnt = counter[col]
        if cnt:
            new_cols.append(f"{col}.{cnt}")
        else:
            new_cols.append(col)
        counter[col] += 1
    return new_cols

# Apply the function to df.columns
df.columns = rename_duplicates(df.columns)
```

prepare final filtered datasets by each level of theory (reindexed):

```python
# Extract columns based on input theory and output bandgap
input_theory_cols = [col for col in df.columns if 'inputs.theory' in col]
output_cols = [col for col in df.columns if col.startswith('outputs.') and col.endswith('.bandgap')]
info_cols = [col for col in df.columns if col.startswith('info.')]
id_cols = [col for col in df if col == 'qmof_id']

# Map inputs to outputs based on order
theory_output_map = dict(zip(input_theory_cols, output_cols))

# Rename inputs based on corresponding outputs
rename_map = {input_col: f"inputs.{output_col.split('.')[1]}.theory" 
               for input_col, output_col in theory_output_map.items()}

# Apply renaming to dataframe
df = df.rename(columns=rename_map)

# Rebuild theory-output map using new names
theory_output_map = dict(zip(rename_map.values(), output_cols))

# Filter DataFrames dynamically
def filter_df(input_col, output_col):
    filtered = df[df[input_col].notna()].copy()
    return filtered[[input_col, output_col] + info_cols + id_cols]

filtered_dfs = {f"df_{input_col.split('.')[1]}": filter_df(input_col, output_col) 
                for input_col, output_col in theory_output_map.items()}

# Reset index for all DataFrames stored in dictionary
filtered_dfs = {key: df.reset_index(drop=True) for key, df in filtered_dfs.items()}

# Access specific DataFrames if needed
df_pbe = filtered_dfs.get('df_pbe')
df_hle17 = filtered_dfs.get('df_hle17')
df_hse06_10hf = filtered_dfs.get('df_hse06_10hf')
df_hse06 = filtered_dfs.get('df_hse06')

# Write all filtered DataFrames to CSV
for key, df in filtered_dfs.items():
    df.to_csv(f"{key}.csv", index=False)
```

**Method 2.** Consider each MOF at each level of theory as an indiviual data point and generate a reformatted dataset as needed for training multi-fidelity models

```python
# Standardize column names before combining
for key, df in filtered_dfs.items():
    theory_name = key.replace('df_', '')  # Safely extract 'pbe', 'hle17', 'hse06_10hf', 'hse06'
    
    input_col = f"inputs.{theory_name}.theory"
    output_col = f"outputs.{theory_name}.bandgap"
    
    # Rename to standard column names
    df = df.rename(columns={input_col: 'inputs.theory', output_col: 'outputs.bandgap'})
    
    # Reset index to clean up and store back
    filtered_dfs[key] = df.reset_index(drop=True)

# Combine all DataFrames into one
df_combined = pd.concat(filtered_dfs.values(), ignore_index=True)

# Write to CSV
df_combined.to_csv('df_combined.csv', index=False)
```
